```python
#Answer 4)
'''SOAP and REST are two internet data exchange mechanisms. SOAP and REST are two different approaches to API design.
The SOAP approach is highly structured and used XML data format. REST is more flexible and allows applications to 
exchange data in multiple formats.

They both describe rules and standards on how applications make, process, and respond to data requests from other
applications.
They both use HTTP, the standardized internet protocol, to exchange information.
They both support SSL/TLS for secure, encrypted communication.

When to use SOAP vs REST?

Overall application design: Modern applications like mobile apps and hybrid applications work better with REST APIs.
However, if you need to integrate or extend lagacy systems that already have SOAP APIs, you may be better off 
continuing with SOAP.

Security: Public APIs have lower security requirements and demand greater flexibility so anyone can interact with them.
So, REST is a better choice when you build public APIs. Conversly, some priavte APIs for internal enterprise requirements
may benefit from the tighter security measures in WS-Security of SOAP.

ACID compliance: SOAP has built-in compliance for atomicity, consistency, isolation, and durability(ACID).And SOAP may
be better suited for high data integrity requirements. In this case, REST APIs may require additional software modules
to enforce the state at the server or database level.

SOAP APIs

SOAP is a protocol that defines rigid communication rules. It has several associated standards that control every aspect
of the data exchange.
    Web Services Security (WS-Security) specifies security measures like using unique identifiers called tokens.
    Web Services Addressing (WS-Addressing) requires including routing information as metadata.
    WS-ReliableMessaging standardizes error handling in SOAP messaging.
    Web Services Description Language (WSDL) describes the scope and function of SOAP web services.

REST APIs

REST is a software architectural style that imposes six conditions on how an API should work.
    Client-server architecture. The sender and receiver are independent of each other regarding technology, platforming,
    programming language, and so on.
    Layered. The server can have several intermediaries that work together to complete client requests, but they are
    invisible to the client.
    Uniform interface. The API returns data in a standard format that is complete and fully uneable.
    Stateless. The API completes every new request independently of previous requests.
    Cacheable. All API responses are cacheable.
    Code on demand. The API responses are cacheable.'''
```

```python
#Answer 5)
'''SOAP stands for Simple Object Access Protocol and REST for Representational State Transfer.

SOAP is a protocol for communication between applications and REST is an architecture style for designing communication
interfaces.

SOAP API exposes the operation whereas REST API exposes the data.

SOAP is independentent and can work with any transport protocol whereas REST works only with HTTPS.

SOAP supports only XML data exchange whereas REST supports XML, JSON, plain text, HTML.

SOAP messages are larger, which makes communication slower. REST has faster performance due to smaller messages and
caching support.

SOAP is difficult to scale. The server maintains state by storing all previous messages exchanged with a client. REST is
easy to scale. Its stateless, so every message is processed independently of previous messages.

SOAP supports encryption with additional overheads. REST supports encryption without affecting performance.

SOAP is useful in legacy applications and private APIs. REST is useful in modern applications and public APIs.'''
```
